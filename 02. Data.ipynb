# KOSPI

```r
krx <- getSymbols('^KS11', from='2017-01-01', to=Sys.Date(), auto.assign=FALSE)
krx.closed <- krx[,'KS11.Close']
par(mfrow = c(2,1))
plot(krx.closed, col = 1)
krx.closed.rate <- diff(log(krx.closed))
plot(krx.closed.rate, col = 2)
```

# Shanghi index

```r
ssec <- getSymbols('000001.SS', from='2017-01-01', to=Sys.Date(), auto.assign=FALSE)
ssec.closed <- ssec[,'000001.SS.Close']
par(mfrow = c(2,1))
plot(ssec.closed, col=3)
ssec.closed.rate <- diff(log(ssec.closed))
plot(ssec.closed.rate, col=4)
```

```r
par(mfrow = c(2,1))
plot(krx.closed, col=1)
plot(ssec.closed, col=3)
```

```r
df <-merge(krx.closed, ssec.closed) 
plot(df, main = "Comparsion between KRX and SSEC")

cor(df[,1], df[,2], use='complete.obs')
```

```r
df2 <-merge(krx.closed.rate, ssec.closed.rate) 
plot(df2, main = "Comparsion between variance of KRX and variance of SSEC")

cor(df2[,1], df2[,2], use='complete.obs')
```
